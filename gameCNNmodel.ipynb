```
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
# from kerastuner import RandomSearch
# from kerastuner.engine.hyperparameters import HyperParameters




import pickle
features=pickle.load(open("/content/drive/My Drive/datasets/gamefeatures.pkl","rb"))
labels=pickle.load(open("/content/drive/My Drive/datasets/gamelabels.pkl","rb"))


import numpy as np
features=features/255.0
labels=np.array(labels).reshape(-1,1)


```

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(features,labels,test_size=0.2,random_state=42)
```

```
def Model_Build(hp):
  model=Sequential()

  model.add(Conv2D(
      filters=hp.Int("Conv2D_1",min_value=30,max_value=300,step=16),
      kernel_size=(3,3),
      activation="relu",
      kernel_initializer="he_uniform",
      input_shape=features.shape[1:]
  ))

  model.add(MaxPooling2D(pool_size=(2,2)))

  for i in range(hp.Int("ADD_layer",1,4)):
    model.add(Conv2D(
        filters=hp.Int(f"Conv_{i}",30,300,15),
        kernel_size=(3,3),
        activation="relu",
        kernel_initializer="he_uniform"
    ))


  model.add(Flatten())
  model.add(Dense(
      units=hp.Int("Dense_units",min_value=64,max_value=256,step=16),
      activation="relu",                                    
  ))
  model.add(Dropout(
      rate=0.3
  ))
  
  for i in range(hp.Int("ADD_DENSE",1,4)):
    model.add(Dense(
        units=hp.Int(f"Add_Dense_units_{i}",min_value=60,max_value=300,step=20),
        activation="relu"
    ))

  model.add(Dense(
      units=3,
      activation="softmax"
  ))
  model.compile(
      optimizer=tf.keras.optimizers.Adam(hp.Choice("learning_rate",values=[1e-2,1e-2,1e-4])),
      loss="sparse_categorical_crossentropy",
      metrics=["accuracy"]
  )

  return model


```

```

random_search=RandomSearch(
    hypermodel=Model_Build,
    objective="val_accuracy",
    max_trials=10,
)
```

```
random_search.search(
    X_train,
    Y_train,
    validation_split=0.12,
    batch_size=30,
    epochs=10
)
```

```
model=random_search.get_best_models(num_models=1)[0]
model.fit(X_train,Y_train,epochs=5,validation_split=0.1,validation_data=(X_test,Y_test))
```

```
results=model.evaluate(X_test,Y_test,batch_size=30)
print(results)
```
